## Tokenization

```python
from transformers import AutoTokenizer

roberta_tokenizer = AutoTokenizer.from_pretrained("FacebookAI/roberta-base")
base_tokenizer=AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")

sentence="TokEnization is great!"

print(sentence)
print("Tokenized sentence with roberta tokenizer",roberta_tokenizer.tokenize(sentence))
print("Tokenized sentence with base tokenizer",base_tokenizer.tokenize(sentence))
```
